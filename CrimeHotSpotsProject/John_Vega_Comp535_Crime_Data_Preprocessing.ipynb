```python
import pandas as pd

# Load the crime.csv file
df = pd.read_csv('Crime_Data_from_2020_to_Present.csv')

# Check the columns to verify the names
print("Columns in CSV file:", df.columns)

# Remove the specified columns
columns_to_remove = ['Cross street', 'Crm Cd 1', 'Crm Cd 2', 'Crm Cd 3', 'Part 1-2', 'Status', 'Rpt dist no']

# Filter out only the columns that exist
columns_to_remove_existing = [col for col in columns_to_remove if col in df.columns]

# Drop the existing columns
df.drop(columns=columns_to_remove_existing, inplace=True)

# Extract only the year from the 'Date Rptd' column if it exists
if 'Date Rptd' in df.columns:
    df['Year'] = pd.to_datetime(df['Date Rptd'], errors='coerce').dt.year

# Save the cleaned DataFrame to a new CSV file
df.to_csv('cleaned_crime.csv', index=False)

# Optional: Print the cleaned dataframe to verify the changes
print(df.head())
```
