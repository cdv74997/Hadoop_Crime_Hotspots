```python
import pandas as pd
# Load CSV
housing_data = pd.read_csv('California_Housing_CitiesAdded.csv')

# Get all the unique city names
unique_city_names = housing_data['City'].unique()


print("\nUnique city names:")
print(unique_city_names)
```

```python
# we are only interested in seeing housing and median income info for the area within the relevant crime district
la_housing_data = housing_data[housing_data['City'] == 'Los Angeles']

# Display the filtered data
print("\nFiltered data for Los Angeles:")
print(la_housing_data)

# Save the filtered data to a new CSV (optional)
la_housing_data.to_csv('LA_Housing_Data_2.csv', index=False)
```

```python
# Count the number of na values in the median house value column
na_count = la_housing_data['Median_House_Value'].isna().sum()

print(f"\nNumber of missing values in 'Median_House_Value': {na_count}")
```

```python
# categorizing the median home value column
bins = [0, 100000, 200000, 300000, 400000, 500000, float('inf')]
labels = ["< $100 thousand", ">= $100 thousand", ">= $200 thousand", ">= $300 thousand", ">= $400 thousand", ">= $Half Million"]
la_housing_data['Median_House_Value_Category'] = pd.cut(la_housing_data['Median_House_Value'], bins=bins, labels=labels)

# Display the filtered and categorized data
print("\nFiltered data for Los Angeles with categorized Median_House_Value:")
print(la_housing_data)

# Save the filtered data to a new CSV (optional)
la_housing_data.to_csv('LA_Housing_Data_Categorized_3.csv', index=False)
```
